```python
import numpy as np
import matplotlib.pyplot as plt

# Data
x = np.linspace(0, 6, 400)
y1 = 3 - 3 * x
y2 = x + 1
y3 = 3 * x + 1.5

# Plot
plt.figure(figsize=(8, 8))
plt.plot(x, y1, label=r"$x_2 = 3 - 3x_1$")
plt.plot(x, y2, label=r"$x_2 = x_1 + 1$")
plt.plot(x, y3, label=r"$x_2 = 3x_1 + 1.5$")

# Feasible region
y_lower = np.maximum(y1, y2)
plt.fill_between(x, y_lower, y3, where=(y3 >= y_lower), alpha=0.3)

# Formatting
plt.xlim(0, 6)
plt.ylim(0, 20)
plt.xlabel(r"$x_1$")
plt.ylabel(r"$x_2$")
plt.title("Feasible Region (Unbounded)")
plt.scatter(0.5, 1.5, color="black", zorder=3)
plt.text(0.5, 1.5, "  (0.5, 1.5)", fontsize=10, va="center")
plt.legend()
plt.grid(True)

# Save and display
plt.savefig("problem1.png", dpi=300, bbox_inches="tight")
plt.show()
```
